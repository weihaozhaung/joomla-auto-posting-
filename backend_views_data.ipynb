```python
from selenium import webdriver
import pandas as pd
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://bigdatafinance.tw/administrator/index.php')
```

# Login BIG DATA

```python
driver.find_element_by_id('mod-login-username').clear()
driver.find_element_by_id('mod-login-username').send_keys('[Username]')
driver.find_element_by_id('mod-login-password').clear()
driver.find_element_by_id ('mod-login-password').send_keys('[Password]')
driver.find_element_by_xpath('//button').click()
```

## -->Article

```python
driver.find_element_by_xpath('//ul[@class="j-links-group nav nav-list"][1]/li[2]/a').click()
```

### 改顯示量

```python
driver.find_element_by_xpath("//div[@id='list_limit_chzn']/a[@class='chzn-single']/span").click()
```

```python
driver.find_element_by_xpath("//div[@class='js-stools-field-list'][2]/div[@id='list_limit_chzn']/div[@class='chzn-drop']/div[@class='chzn-search']/input").send_keys('所有')
```

```python
driver.find_element_by_xpath("//li[@class='active-result highlighted']").click()
```

```python
Textdata = pd.DataFrame()
```

```python
i = 1
while True:
    try:
        title=driver.find_element_by_xpath("//tr[@class='row0'][%s]//div[@class='pull-left break-word']/a[@class='hasTooltip']"%i)
    except:
        print('finished')
        break
    Date=driver.find_element_by_xpath("//tr[@class='row0'][%s]/td[@class='nowrap small hidden-phone']"%i)
    Classification1 = driver.find_element_by_xpath("//tbody/tr[@class='row0'][%s]//div[@class='small']/a[@class='hasTooltip'][1]"%i)
    try:
         Classification2 = driver.find_element_by_xpath("//tbody/tr[@class='row0'][%s]//div[@class='small']/a[@class='hasTooltip'][2]"%i).text
    except:
         Classification2 = "NA"
    View=driver.find_element_by_xpath("//tr[@class='row0'][%s]/td[@class='hidden-phone center'][1]/span[@class='badge badge-info']"%i)
    Temp = pd.DataFrame({"Title" : [title.text],"Date" : [Date.text],"Type1": [Classification1.text],"Type2" : [Classification2],"PageView" : [View.text] })
    Textdata=Textdata.append(Temp)
    try:
        title=driver.find_element_by_xpath("//tr[@class='row1'][%s]//div[@class='pull-left break-word']/a[@class='hasTooltip']"%i)
    except:
        print('finished')
        break
        
    Date=driver.find_element_by_xpath("//tr[@class='row1'][%s]/td[@class='nowrap small hidden-phone']"%i)
    try:
         Classification2 = driver.find_element_by_xpath("//tbody/tr[@class='row0'][%s]//div[@class='small']/a[@class='hasTooltip'][2]"%i).text
    except:
         Classification2 = "NA"
    View=driver.find_element_by_xpath("//tr[@class='row1'][%s]/td[@class='hidden-phone center'][1]/span[@class='badge badge-info']"%i)
    Temp = pd.DataFrame({"Title" : [title.text],"Date" : [Date.text],"Type1": [Classification1.text],"Type2" : [Classification2],"PageView" : [View.text] })
    Textdata=Textdata.append(Temp)
    print("剩餘:"+str(int(driver.find_element_by_xpath("//tr[@class='row0'][%s]/td[@class='hidden-phone']"%i).text)-74))
    i += 1
```
